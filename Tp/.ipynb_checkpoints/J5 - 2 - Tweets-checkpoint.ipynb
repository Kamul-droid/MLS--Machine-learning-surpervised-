```python
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train = pd.read_pickle("../J4/X_train.pkl")
X_test = pd.read_pickle("../J4/X_test.pkl")
y_train = pd.read_pickle("../J4/y_train.pkl")
y_test = pd.read_pickle("../J4/y_test.pkl")
```

```python
X_train
```

```python
random_forest = RandomForestClassifier()
```

```python
%%time
random_forest.fit(X_train, y_train)
```

```python
y_pred = random_forest.predict(X_test)
```

```python
print(classification_report(y_pred, y_test))
```

Attention : `random_forest.feature_names_in_` et `random_forest.feature_importances_` ne semble plus disponible sur scikit-learn > 1.3.0

```python
importances = pd.DataFrame({
    "variable" : random_forest.feature_names_in_,
    "importance": random_forest.feature_importances_
})
```
