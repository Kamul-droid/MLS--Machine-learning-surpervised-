# Prétraitement

Méthode pour détecter les anomalies numériques

- Calculer les quartiles : Q1 et Q3
- Calculer la distance interquartile : delta = Q3-Q1
- Valeurs à garder : [Q1 -1.5 x delta ; Q3 + 1.5 x delta]

````python
class OutlierRemover(BaseEstimator, TransformerMixin):
    def __init__(self, columns, factor = 1.5, remove=True):
        self.columns = columns
        self.factor = factor
        self.remove = remove
        self.kept_indices = None
        
    def remove_outliers(self, X):
        l = []
        for c in self.columns:
            Q1 = X[c].quantile(0.25)
            Q3 = X[c].quantile(0.75)
            delta = Q3-Q1
            lower_bound = Q1 - self.factor*delta
            upper_bound = Q3 + self.factor*delta
            column = X[c].loc[ ( (X[c] >= lower_bound) & (X[c] <= upper_bound) )  ]
            l.append( pd.Series(column) )
            
        res = pd.concat(l, axis=1, join = "inner")
        
        self.kept_indices = res.index
        
        return X.loc[self.kept_indices]
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        if self.remove:
            return self.remove_outliers(X)
        else:
            self.kept_indices = X.index
            return X
````

````python
outlier_pipeline = Pipeline(steps = [("outliers", OutlierRemover(columns=num_columns, factor=1.5, remove=False))])
````

````python
color_pipeline = Pipeline(steps = [("encoder", OrdinalEncoder(categories=[["J", "I", "H", "G", "F", "E", "D"]]))])
cut_pipeline = Pipeline(steps = [("encoder", OrdinalEncoder(categories=[["Fair", "Good", "Very Good", "Premium", "Ideal"]]))])
clarity_pipeline = Pipeline(steps = [("encoder", OrdinalEncoder(categories=[["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]]))])
````

````python
num_preprocessor = ColumnTransformer([
    ("scaler", MinMaxScaler(), num_columns)
], 
    remainder="passthrough",
    verbose_feature_names_out=False
)
num_preprocessor.set_output(transform="pandas")
````

````python
ord_preprocessor = ColumnTransformer([
    ("color", color_pipeline, ["color"]),
    ("cut", cut_pipeline, ["cut"]),
    ("clarity", clarity_pipeline, ["clarity"]),
], 
    remainder="passthrough",
    verbose_feature_names_out=False
)
ord_preprocessor.set_output(transform="pandas")
````

````python
pipeline = Pipeline(steps=[
    ("outliers", outlier_pipeline),
    ("num", num_preprocessor),
    ("ord", ord_preprocessor)
],verbose = 4)
````

````python
pipeline
````

````python
X_train_clean = pipeline.fit_transform(X_train)
y_train_clean = y_train.loc[pipeline.named_steps["outliers"].named_steps["outliers"].kept_indices]
````

````python
X_test_clean = pipeline.transform(X_test)
y_test_clean = y_test.loc[pipeline.named_steps["outliers"].named_steps["outliers"].kept_indices]
````

````python
print(X_train_clean.shape)
#print(X_val.shape)
print(X_test_clean.shape)
print(y_train_clean.shape)
#print(y_val.shape)
print(y_test_clean.shape)
````

# Régression linéaire

````python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.linear_model import LinearRegression
````

````python
pipeline_final = Pipeline(steps = [
    ("preprocessing", pipeline),
    ("linear_regression", LinearRegression())
])
````

````python
pipeline_final.fit(X_train, y_train_clean)
````

````python
y_pred = pipeline_final.predict(X_test)
````

````python
y_pred
````

````python
print("Mean Squared Error      : ", mean_squared_error(y_test_clean, y_pred))
print("Root Mean Squared Error : ", math.sqrt(mean_squared_error(y_test_clean, y_pred)))
print("Mean Absolute Error     : ", mean_absolute_error(y_test_clean, y_pred))
print("R2                      : ", r2_score(y_test_clean, y_pred))
````

````python
def plot_pred_vs_test(y_pred, y_test):
    plt.scatter(y_test, y_pred, s=1)
    maxi = max(y_test.max(), y_pred.max())
    #plt.xlim(0, maxi)
    #plt.ylim(0, maxi)
    ax = plt.gca()
    ax.set_aspect('equal', adjustable='box')
    plt.xlabel("Real price ($)")
    plt.ylabel("Estimated price ($)")
    plt.plot([0,maxi], [0, maxi], color = "r", ls="--", linewidth=1)
    plt.show()
````

````python
plot_pred_vs_test(y_pred, y_test_clean)
````

````python
def plot_pred_vs_test_relative(y_pred, y_test):
    plt.scatter(y_test, 100*(y_pred-y_test)/y_test, s=1)
    maxi = y_test.max()
    plt.xlim(0, maxi)
    plt.xlabel("Real price ($)")
    plt.ylabel("Relative error (%)")
    plt.plot([0, maxi], [0, 0], color = "r", ls="--", linewidth=1)
    plt.show()
````

````python
plot_pred_vs_test_relative(y_pred, y_test_clean)
````

# Régression avec les k plus proches voisins

````python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
````

````python
pipeline_final_knn = Pipeline(steps = [
    ("preprocessing", pipeline),
    ("knn", KNeighborsRegressor())
])
````

````python
hyperparameters_knn = {
    "preprocessing__outliers__outliers__remove": [False],
    "knn__n_neighbors" : range(2,10),
    "knn__weights" : ["uniform", "distance"],
}
````

GridSearchCV vise à maximiser la fonction de score. Dans un problème de régression, on souhaite minimiser l'erreur (MSE ou MAE). Il faut donc renseigner un scoring égal à l'opposé de l'erreur : "neg_mean_squared_error" ou "neg_mean_absolute_error"

````python
gs_knn = GridSearchCV(
    estimator=pipeline_final_knn,
    param_grid=hyperparameters_knn,
    cv = 4,
    scoring = "neg_mean_squared_error",
    verbose = 10
)
````

````python
gs_knn.fit(X_train, y_train)
````

````python
knn = gs_knn.best_estimator_
````

````python
gs_knn.best_params_
````

````python
y_pred = knn.predict(X_test)
````

````python
print("Mean Squared Error      : ", mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error : ", math.sqrt(mean_squared_error(y_test, y_pred)))
print("Mean Absolute Error     : ", mean_absolute_error(y_test, y_pred))
print("R2                      : ", r2_score(y_test, y_pred))
````

````python
plot_pred_vs_test(y_test,y_pred)
````

````python
plot_pred_vs_test_relative(y_test,y_pred)
````
