```python
import pandas as pd
import numpy as np
from sklearn.svm import LinearSVC, SVC
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    precision_score,
    recall_score,
    classification_report,
)
from sklearn.model_selection import GridSearchCV
```

```python
X_train = pd.read_csv("../J1/X_train_clean.csv")
X_test = pd.read_csv("../J1/X_test_clean.csv")
y_train = pd.read_csv("../J1/y_train.csv")
y_test = pd.read_csv("../J1/y_test.csv")
```

# Classification avec LinearSVC

```python
import numpy as np
```

```python
start_c = 0.233
end_c = 0.235
step_c = 0.00001

hyperparameters = [
    {
        "penalty" : ["l1", "l2"],
        "loss": ["squared_hinge"],
        "dual": [False],
        "max_iter": [10000],
        "C": np.arange(start_c, end_c, step=step_c),
    }
]
```

```python
gscv_linear_svm = GridSearchCV(
    estimator = LinearSVC(),
    param_grid = hyperparameters,
    cv = 5,
    scoring = "f1",
    verbose = 4
)
```

```python
gscv_linear_svm.fit(X_train, y_train.values.ravel())
```

```python
best_linear_svm = gscv_linear_svm.best_estimator_
```

```python
best_linear_svm_hyperparameters = gscv_linear_svm.best_params_
```

```python
best_linear_svm_hyperparameters
```

```python
y_pred = best_linear_svm.predict(X_test)
```

```python
print("Accuracy  :", accuracy_score(y_pred, y_test.values))
print("Precision :", precision_score(y_pred, y_test.values))
print("Recall    :", recall_score(y_pred, y_test.values))
print("F1        :", f1_score(y_pred, y_test.values))
```

```python
print("Accuracy  :", accuracy_score(y_pred, y_test.values))
print("Precision :", precision_score(y_pred, y_test.values))
print("Recall    :", recall_score(y_pred, y_test.values))
print("F1        :", f1_score(y_pred, y_test.values))
```

```python
print(classification_report(y_pred, y_test))
```

# Classification avec SVC

```python
start_c = 0.1
end_c = 5
step_c = 0.5

start_gamma = 0.5
end_gamma = 0.6
step_gamma = 0.01

hyperparameters = [
    {
        "C": np.arange(start_c, end_c, step=step_c),
        "kernel": ["poly", "rbf", "sigmoid"],
        "gamma": np.arange(start_gamma, end_gamma, step=step_gamma),
        
    }
]
```

```python
gscv_svm = GridSearchCV(
    estimator = SVC(),
    param_grid = hyperparameters,
    cv = 5,
    scoring = "f1",
    verbose = 4
)
```

```python
gscv_svm.fit(X_train, y_train.values.ravel())
```

```python
best_svm = gscv_svm.best_estimator_
```

```python
best_svm_hyperparameters = gscv_svm.best_params_
```

```python
best_svm_hyperparameters
```

```python
y_pred = best_svm.predict(X_test)
```

```python
print("Accuracy  :", accuracy_score(y_pred, y_test.values))
print("Precision :", precision_score(y_pred, y_test.values))
print("Recall    :", recall_score(y_pred, y_test.values))
print("F1        :", f1_score(y_pred, y_test.values))
```

```python
print("Accuracy  :", accuracy_score(y_pred, y_test.values))
print("Precision :", precision_score(y_pred, y_test.values))
print("Recall    :", recall_score(y_pred, y_test.values))
print("F1        :", f1_score(y_pred, y_test.values))
```

```python
print(classification_report(y_pred, y_test))
```
